```python
import pandas as pd
data={
    "Department":["IT","IT","HR","HR","FINANCE"],
    "Name":["A","B","C","D","E"],
    "Marks":[80,90,75,99,88]
}
df=pd.DataFrame(data)
print("original data:")
print(df)
print("\n avg marks by dept")
print(df.groupby("Department")["Marks"].mean())
print("\n total marks by dept")
print(df.groupby("Department")["Marks"].sum())
print("\n student count by dept:")
print(df.groupby("Department")["Marks"].count())
```

```python
#rearranging
import pandas as pd

data = {
    "Name": ["A", "B", "C"],
    "Age": [22, 21, 23],
    "Marks": [80, 90, 80]
}

df = pd.DataFrame(data)

print("Original Data Frame:")
print(df)

# Rearranging rows in ascending order based on Marks
df_sorted = df.sort_values(by="Marks")
print("\nAfter rearranging rows (ascending):")
print(df_sorted)

# Rearranging columns
df_rearranged_cols = df[["Name", "Age", "Marks"]]
print("\nAfter rearranging columns:")
print(df_rearranged_cols)
```

```python
#ranking
import pandas as pd
data={
    "Name":["A","B","C","D"],
    "Marks":[85,92,78,92]
}
df=pd.DataFrame(data)
print("original data frame:")
print(df)
df["Rank"]=df["Marks"].rank(ascending=False)
print("\n after ranking")
print(df)
```
