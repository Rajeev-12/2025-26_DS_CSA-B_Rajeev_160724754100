```python
#binary search
n=int(input("enter no. of elements"))
arr=[]
for i in range(n):
  arr.append(int(input(f"enter element{i+1}:")))
arr.sort()
print("\n sorted list",arr)
key=int(input("enter element to search:"))
low=0
high=n-1
found=False
while low<=high:
  mid=(low+high)//2
  if arr[mid]==key:
    print(f"element {key} found position {mid+1}")
    found=True
    break
  elif (key<arr[mid]):
      high=mid-1
  else:
      low=mid+1
  if not found:
    print("element not found")
```
